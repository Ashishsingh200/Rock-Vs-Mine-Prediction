```python
#importing the Library
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#getting the datasets
sonar_data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning/Copy of sonar data.csv',header=None)
```

```python
#seperating the data and the labels
x=sonar_data.drop(60,axis=1)
y=sonar_data[60]
```

```python
x
```

```python
y
```

```python
#traning and testing the data set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
#stratify means data will be splitted based on r and m
```

```python
print(x.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#Model training
model=LogisticRegression()
```

```python
#training model
model.fit(x_train , y_train)
```

```python
#model evaluation
#finding accuracy on the training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on training data :',training_data_accuracy)
```

```python
#finding accuracy on the test data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy of test data :',testing_data_accuracy)
```
